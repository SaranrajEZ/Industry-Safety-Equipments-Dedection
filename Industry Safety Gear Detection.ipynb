## Load YOLOv8 Model & Train the Model



```python

# Load the YOLOv8 model
model = YOLO('yolov8s.pt')

# Train the model
model.train(data="/content/drive/MyDrive/DL project/Industry Safety Gear dedection/data/data.yaml", epochs=25, imgsz=224, plots=True)
```
